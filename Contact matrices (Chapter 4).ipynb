```python
# importing relevant libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# All contacts

# Data available from Prem et al. 2017, although their updated 2021 paper is recommended

df_all = pd.read_excel(r"C:", usecols = 'B:Q') # read in
All = df_all.iloc[2:18] # take desired rows
All = All.astype(float) # convert to float for plotting purposes

axticks = [5+5*col for col in range(16)] # relabel axes with ages 5-80 rather than 1-16

fig, ax = plt.subplots(figsize=(2.8,2.1)) # small figsize ensures readable labels
ax = sns.heatmap(All.T, cmap = 'Purples', xticklabels = axticks, yticklabels = axticks, cbar_kws={'label': 'Number of daily contacts'})
ax.invert_yaxis() # transpose and reflect to effectively rotate the heatmap by 90° (reads easier)     
ax.axhline(0, color = 'k', linewidth = 3)
ax.axvline(0, color = 'k', linewidth = 3) # highlight the axes
ax.set(xlabel = 'Age of individual', ylabel = 'Age of contact', title = '"All" contacts')

new_ticks = [0]+[i.get_text() for i in ax.get_xticklabels()]
plt.xticks(range(0, len(new_ticks), 2), new_ticks[::2])
plt.yticks(range(0, len(new_ticks), 2), new_ticks[::2]) # show every other tick (multiples of 10)
plt.xticks(rotation=40)
plt.yticks(rotation=0) # avoid 'vertical' text
```
